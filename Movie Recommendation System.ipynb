```python
def conversion(x):
    genre_list=[]
    x=ast.literal_eval(x)
    for genre in x:
        genre_list.append(genre['name'])
    return genre_list
```

```python
def convert_cast(x):
    list_cast=[]
    x=ast.literal_eval(x)
    counter=0
    for i in x:
        if counter <= 3:
            list_cast.append(i['name'])
            counter+=1
        else:
            break
    return list_cast
```

```python
df.cast=df.cast.apply(convert_cast)
```

```python
def convert_crew(obj):
    list_director=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list_director.append(i['name'])
        else:
            pass
    return list_director
```

```python
df.crew=df.crew.apply(convert_crew)
```

```python
df.crew=df.crew.apply(lambda x:[i.replace(" ","") for i in x])
df.cast=df.cast.apply(lambda x:[i.replace(" ","") for i in x])
df.genres=df.genres.apply(lambda x:[i.replace(" ","") for i in x])
df.keywords=df.keywords.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head(5)
```

```python
df['tags']=df.overview+df.genres+df.keywords+df.cast+df.crew
```

```python
new_df=df[['movie_id','title','tags']]
```

```python
new_df.tags=new_df.tags.apply(lambda x: " ".join(x))
```

```python
new_df.tags[0]
```

```python
new_df.tags=new_df.tags.apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000)
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
def convert_lemma(text):
    conv_lemm_list=[]
    for i in text.split():
        conv_lemm_list.append(lemma.lemmatize(i))
    return " ".join(conv_lemm_list)
```

```python
new_df.tags=new_df.tags.apply(convert_lemma)
```

```python
vector_lemma_list=cv.fit_transform(new_df.tags).toarray()
```

```python
cv.get_feature_names()[0:100]
```

```python
vector_lemma_list[0][0:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
SimilarityScore=cosine_similarity(vector_lemma_list)
```

```python
def Recommend(movie_name):
    movie_index=df[df.title == movie_name].index[0]
    movie_top_10=sorted(list(enumerate(SimilarityScore[movie_index])),reverse=True,key=lambda x:x[1])[1:10]
    distance=[]
    recommended=[]
    for i in movie_top_10:
        recommended.append(new_df.iloc[i[0]].title)
        distance.append(i[1])
    return recommended,distance
```
